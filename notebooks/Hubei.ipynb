## Type date columns

```python
import pandas as pd

for column in ncov_hubei.columns:
    if column.startswith("date"):
        ncov_hubei[column] = pd.to_datetime(ncov_hubei[column], errors="coerce", dayfirst=True)
```

## Drop rows where confirmation date is after death or discharge date

None exist; added for consistency with `Outside Hubei.ipynb`.

```python
ncov_hubei[ncov_hubei.date_confirmation > ncov_hubei.date_death_or_discharge]
```

```python
ncov_hubei_filtered = ncov_hubei[
    ~(ncov_hubei.date_confirmation > ncov_hubei.date_death_or_discharge)
    & ncov_hubei.date_confirmation.notna()
]
```

# Get data in the right format

```python
start_dates = ncov_hubei_filtered.date_confirmation
end_dates = ncov_hubei_filtered.date_death_or_discharge
```

```python
from lifelines.utils import datetimes_to_durations

T, E = datetimes_to_durations(start_dates, end_dates, freq="D")
pd.DataFrame({"T": T, "E": E})[E]
```
